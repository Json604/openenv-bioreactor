# BioOperatorEnv — Smoke Test

Run a few episodes of the env to verify reset/step/reward all behave. No GPU needed.

```python
import sys; sys.path.insert(0, '..')
from bioperator_env.env import BioOperatorEnv
from bioperator_env.scenarios import list_tasks
print('available tasks:', list_tasks())
```

```python
env = BioOperatorEnv(task_id='do-recovery-medium', seed=42)
obs = env.reset()
print('reset OK at t_h =', obs.time_h, 'phase =', obs.batch_phase)
print('alarm:', obs.alarm)
print('measurements:')
for k, v in obs.measurements.items():
    print(f'  {k}: {v:.3f}' if isinstance(v, float) else f'  {k}: {v}')
```

```python
from baselines.rule_based_agent import RuleBasedAgent
agent = RuleBasedAgent()
obs = env.reset()
rewards = []
done = False
while not done:
    a = agent.act(obs)
    obs, r, done, info = env.step(a)
    rewards.append(r)
print('episode steps:', len(rewards))
print('cumulative reward:', sum(rewards))
print('safety violations:', env.safety_violations)
print('done reason:', info['done_reason'])
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(8,3))
plt.plot(rewards, marker='.')
plt.xlabel('step'); plt.ylabel('reward')
plt.title('Per-step reward (rule-based agent on do-recovery-medium)')
plt.grid(alpha=0.3)
plt.show()
```
